```r
library(sf)
library(rstac)
library(dplyr)
```

# Set constants
<hr style="border:1px solid #0077b9;">

```r
HARMONIZE_STAC_URL <- 'https://brazildatacube.dpi.inpe.br/harmonize/dev/stac/v1/'

FINAL_CSV_PREFIX <- "data_36_months_"

START_DATE <- "2019-01-01"
END_DATE <- "2021-12-31"
```

# List Harmonize STAC collections (using RSTAC)
<hr style="border:1px solid #0077b9;">

```r
all_collections <- stac(HARMONIZE_STAC_URL) %>%
  collections() %>%
  get_request()

for (collection in all_collections$collections){
  print(collection$id)
}
```

# Obtain dengue data
<hr style="border:1px solid #0077b9;">

### Set collection id

```r
DENGUE_COLLECTION_ID <- "dengue_cases_mun_month_north-1"
```

### Search Dengue data on Harmonize STAC (Using RSTAC)

```r
dengue_search <- stac(HARMONIZE_STAC_URL) %>%
  stac_search(
    collections = DENGUE_COLLECTION_ID,
    datetime = paste(START_DATE, END_DATE, sep = "/"),
    limit = 50
  ) %>%
  get_request()

dengue_search$context
```

### For each feature, retrieve geojson as dataframe

```r
dengue_feature_to_geojson_csv <- function(feature) {
  datetime <- feature$properties$datetime
  geojson_href <- feature$assets$geojson$href

  df <- as.data.frame(st_read(geojson_href)) %>%
    dplyr::select(cod, value) %>%
    rename(cd_mun = cod, cases = value) %>%
    mutate(cases = as.numeric(cases)) %>%
    mutate(year_month = format(as.Date(datetime), "%Y-%m"))
}

dengue_dfs <- lapply(X = dengue_search$features, FUN = dengue_feature_to_geojson_csv)
dengue_dfs[1:2]
```

### Rbind lists of monthly dataframes in one big dataframe

```r
dengue_df <- do.call(rbind, dengue_dfs)

dim(dengue_df)
head(dengue_df)
```
