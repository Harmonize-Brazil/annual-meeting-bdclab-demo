```r
library(sf)
library(rstac)
library(dplyr)
```

# Set constants
<hr style="border:1px solid #0077b9;">

```r
HARMONIZE_STAC_URL <- 'https://brazildatacube.dpi.inpe.br/harmonize/dev/stac/v1/'

FINAL_CSV_PREFIX <- "data_2020_week_"

START_DATE <- "2020-01-01"
# END_DATE <- "2021-12-31"
END_DATE <- "2020-12-31"
```

# List Harmonize STAC collections (using RSTAC)
<hr style="border:1px solid #0077b9;">

```r
all_collections <- stac(HARMONIZE_STAC_URL) %>%
  collections() %>%
  get_request()

for (collection in all_collections$collections){
  print(collection$id)
}
```

# Obtain monthly climate data for each municipality
<hr style="border:1px solid #0077b9;">

### Set collection id (temperature, precipitation, etc)

```r
# CLIMATE_COLLECTION_ID <- "max_temper_mean_no_mun_month-1"
# CLIMATE_COLLECTION_ID <- "total_prec_max_no_mun_month-1"
# CLIMATE_COLLECTION_ID <- "total_prec_max_no_mun_epiweek-1"
CLIMATE_COLLECTION_ID <- "max_temper_mean_no_mun_epiweek-1"

# DATA_TYPE <- "precipitation"
DATA_TYPE <- "temperature"

# [1] "total_prec_max_no_mun_month-1"
# [1] "total_prec_mean_no_mun_month-1"
# [1] "total_prec_mean_ne_mun_month-1"
# [1] "total_prec_max_ne_mun_month-1"
```

### Search climate data on Harmonize STAC (Using RSTAC)

```r
climate_search <- stac(HARMONIZE_STAC_URL) %>%
  stac_search(
    collections = CLIMATE_COLLECTION_ID,
    datetime = paste(START_DATE, END_DATE, sep = "/"),
    limit = 200
  ) %>%
  get_request()

climate_search$context
```

### For each feature, retrieve geojson as dataframe

```r
climate_feature_to_geojson_csv <- function(feature) {
  datetime <- feature$properties$datetime
  geojson_href <- feature$assets$geojson$href
  print(geojson_href)

  as.data.frame(st_read(geojson_href)) %>%
    mutate(date = format(as.Date(datetime), "%Y-%m-%d"))
}

climate_dfs <- lapply(X = climate_search$features, FUN = climate_feature_to_geojson_csv)
climate_dfs[1:2]
```

### Rbind lists of monthly dataframes in one big dataframe

```r
climate_df <- do.call(rbind, climate_dfs) %>%
  dplyr::select(CD_MUN, NM_MUN, value, date) %>%
  rename(cd_mun = CD_MUN, nm_mun = NM_MUN, !!DATA_TYPE := value)

dim(climate_df)
head(climate_df, 50)
```

### Save climate data

```r
FILENAME <- paste0(FINAL_CSV_PREFIX, DATA_TYPE, ".csv")

write.csv(climate_df, FILENAME, row.names = FALSE)
```
